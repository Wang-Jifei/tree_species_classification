# Watershed segmentation
### Input: chm.tif
### Output: treetops location; watershed segmentation image

```python
# import required packages
import numpy as np
import gdal
import ogr
from skimage import exposure
from skimage.segmentation import quickshift, slic, watershed
import geopandas as gpd
import numpy as np
from skimage.feature import peak_local_max
import time
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
print("import successrully!")
```

```python
naip_fn = 'data/clip3.tif' # image dir
driverTiff = gdal.GetDriverByName('GTiff') # image type
naip_ds = gdal.Open(naip_fn) # open image
nbands = naip_ds.RasterCount # count bands
band_data = [] # save band data

chm_fn = 'data/clip3.tif' # chm dir
chm_ds = gdal.Open(chm_fn) # open chm image
chm = chm_ds.GetRasterBand(1).ReadAsArray() # image to array
 
for i in range(1, nbands + 1): # loop through all bands
    band = naip_ds.GetRasterBand(i).ReadAsArray() # image to array
    band_data.append(band) # add band
band_data = np.dstack(band_data) # stack bands
img = exposure.rescale_intensity(band_data) # rescale array

# do segmentation, different options with quickshift and slic (only use one of the next two lines)
# segments = quickshift(img.astype('double'), kernel_size=10, ratio=0.8, max_dist=10000, convert2lab=False,random_seed=42)
# segments = slic(img.astype('double'), n_segments=1000, compactness=0.1,max_iter=100)

# local maxima
image_max = ndi.maximum_filter(chm, size=20, mode='constant')

# save treetop coordinates
coor_30 = peak_local_max(chm, min_distance=30) # min_distance=30
coor_40 = peak_local_max(chm, min_distance=40) # min_distance=40
coor_50 = peak_local_max(chm, min_distance=50) # min_distance=50
coor_60 = peak_local_max(chm, min_distance=60) # min_distance=60

# set figure size
fig, axes = plt.subplots(ncols=4, figsize=(12, 3), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(chm) # show chm
ax[0].autoscale(False) # autoscale false
x=coor_30[:, 1] # x_coor
y=coor_30[:, 0] # y_coor
ax[0].plot(x,y, 'r.') # plot points
ax[0].set_title('Distance=30') # set title

ax[1].imshow(chm)# show chm
ax[1].autoscale(False)#  autoscale false
x=coor_40[:, 1] # x_coor
y=coor_40[:, 0]# y_coor
ax[1].plot(x,y, 'r.')# plot points
ax[1].set_title('Distance=40') # set title

ax[2].imshow(chm)# show chm
ax[2].autoscale(False)# autoscale false
x=coor_50[:, 1]# x_coor
y=coor_50[:, 0]# y_coor
ax[2].plot(x,y, 'r.')# plot points
ax[2].set_title('Distance=50') # set title

ax[3].imshow(chm)# show chm
ax[3].autoscale(False)# autoscale false
x=coor_60[:, 1]# x_coor
y=coor_60[:, 0]# y_coor
ax[3].plot(x,y, 'r.')# plot points
ax[3].set_title('Distance=60') # set title
#ax[1].imshow(segments, cmap=plt.cm.nipy_spectral)
#ax[1].set_title('Separated objects')
plt.savefig('figures/treetops.png')# save figure
plt.show()# show figure


```

```python

seg_start=time.time() # set start time
treetops_40 =peak_local_max(chm, min_distance=40,indices=False,footprint=np.ones((3, 3)), labels=None) # detect treetops distance=40
treetops_50 =peak_local_max(chm, min_distance=100,indices=False,footprint=np.ones((3, 3)), labels=None)# detect treetop distance=50
markers_40 = ndi.label(treetops_40)[0]
markers_50 = ndi.label(treetops_50)[0]
# perform watershed segmentation
segments_40 = watershed(chm.astype('double'),markers=markers_40, connectivity=1, offset=None, mask=None, compactness=0, watershed_line=False)
segments_50 = watershed(chm.astype('double'),markers=markers_50, connectivity=1, offset=None, mask=None, compactness=0,watershed_line=False)
#segments_60 = watershed(chm.astype('double'),markers=markers, connectivity=1, offset=None, mask=None, compactness=0, watershed_line=False)

# print total time
print('segments complete in', time.time() - seg_start, 'seconds.')

fig, axes = plt.subplots(ncols=2, figsize=(6, 3), sharex=True, sharey=True)
#plt.suptitle("Watershed segmentation")
ax = axes.ravel()
ax[0].imshow(segments_40, cmap=plt.cm.nipy_spectral)

x=coor_40[:, 1]
y=coor_40[:, 0]
ax[0].plot(x,y, 'r.')
ax[0].set_title('Segment Distance=40')

ax[1].imshow(segments_50, cmap=plt.cm.nipy_spectral)
ax[1].autoscale(False)
x=coor_50[:, 1]
y=coor_50[:, 0]
ax[1].plot(x,y, 'r.')
ax[1].set_title('Segment Distance=50')

plt.savefig('figures/waterseg.png') # save figure
plt.show() # show figure
# Save segmentation results to rasters
segments_fn = 'results/seg_watershed.tif' # segmented image dir
segments_ds = driverTiff.Create(segments_fn, naip_ds.RasterXSize, naip_ds.RasterYSize,
                                1, gdal.GDT_Float32) # save image as tiff

segments_ds.SetGeoTransform(naip_ds.GetGeoTransform()) # transform
segments_ds.SetProjection(naip_ds.GetProjectionRef()) # set projection
segments_ds.GetRasterBand(1).WriteArray(segments_50) # array to image
segments_ds = None
```
